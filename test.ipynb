# Tests

## Test 0

```python
regex_pattern0 = "a[^bc]d*e"
regex_compiler0 = RegexFSM(regex_pattern0)
with open("examples/pattern0.dot", "w+", encoding="utf8") as file:
    file.write(regex_compiler0.to_dot_file())
assert regex_compiler0.check_string("afde") is True
assert regex_compiler0.check_string("azzzze") is False
assert regex_compiler0.check_string("abde") is False      
assert regex_compiler0.check_string("acde") is False
assert regex_compiler0.check_string("agde") is True
assert regex_compiler0.check_string("agggge") is False
with open("examples/pattern0.dot", "r") as file:
    dot_code = file.read()

Source(dot_code)
```
